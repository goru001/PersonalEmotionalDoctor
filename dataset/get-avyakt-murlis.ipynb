```python
# Complete thing
baseUrl = 'http://www.bkdrluhar.com'
englishAvyaktMurlisUrls = []
for url in englishIndexPage:
    html_doc = ''
    with urlopen(url) as response:
        for line in response:
            line = line.decode('utf-8')
            html_doc = html_doc + line.replace('\n','')
    soup = BeautifulSoup(html_doc, 'html.parser')
    print(soup.title.string)
    table = (soup.find('table')).find('table')
    rows = table.find_all('tr')
    print(len(rows))
    for row in rows:
        tds = row.find_all('td')
        htmlUrl = ''; pdfUrl = ''; mp3Url = ''; mp4Url = ''
        if tds[2].find('a') and tds[2].find('a')['href']:
            htmlUrl = baseUrl + tds[2].find('a')['href'][5:]
        if tds[3].find('a') and tds[3].find('a')['href']:
            pdfUrl = baseUrl + tds[3].find('a')['href'][5:]
        if tds[4].find('a') and tds[4].find('a')['href']:
            mp3Url = baseUrl + tds[4].find('a')['href'][5:]
        if tds[5].find('a') and tds[5].find('a')['href']:
            mp4Url = baseUrl + tds[5].find('a')['href'][5:]
        if htmlUrl == '' and pdfUrl == '' and mp3Url == '' and mp4Url == '':
            continue
        englishAvyaktMurlisUrls.append([htmlUrl, pdfUrl, mp3Url, mp4Url])
    print(len(englishAvyaktMurlisUrls))
```

```python
url = 'http://www.bkdrluhar.com/00-Avyakt%20Murli/02.%20English/English%20htm/1969/01.%20AV-E-18.01.1969.htm'
```

```python
html_doc = ''
with urlopen(url) as response:
    for line in response:
        line = line.decode('utf-8')
        html_doc = html_doc + line.replace('\n','')
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.title.string)
```

```python
paras = soup.find_all('p')
```

```python
paras[5]
```

```python
def cleanTextEnglish(text):
    text = text.replace('\r\t\t\t','')
    text = text.replace('  ',' ')
    text = text.replace('\t\t\t\t\t','')
    text = text.replace('\t','')
    text = text.replace(u'\ufeff','')
    text = text.replace(u'\xa0', u' ')
    text = text.replace(u'\x96', u'-')
    return text
```

```python
for para in paras:
    print(cleanTextEnglish(para.text))
```

```python
# Getting all murli texts
englishAvyaktMurlis = []
for i, urls in enumerate(englishAvyaktMurlisUrls):
    url = urls[0]
    if url == '':
        continue
    html_doc = ''
    with urlopen(url) as response:
        for line in response:
            line = line.decode('utf-8', errors='ignore')
            html_doc = html_doc + line.replace('\n','')
    soup = BeautifulSoup(html_doc, 'html.parser')
    print(i)
    paras = soup.find_all('p')
    murli = ''
    for para in paras:
        murli += cleanTextEnglish(para.text)
    englishAvyaktMurlis.append(murli)
```

```python
len(englishAvyaktMurlis)
```

```python
with open('englishAvyaktMurlis.pkl', 'wb') as f:
    pickle.dump(englishAvyaktMurlis,f)
```
